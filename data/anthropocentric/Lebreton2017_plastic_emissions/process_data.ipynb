```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('source/Lebreton2017_plastic_emissions.xlsx')
data.head()
```

```python
df = data.set_index(keys='River')
df.head()
```

```python
df_stacked = df.stack()
df_stacked[0:10]
```

```python
df_new = df_stacked.reset_index()
df_new.head()
```

```python
df_new = df_new.rename(columns={'level_1': 'measurement', 0: 'value'})
df_new.head()
```

```python
df_new.to_csv('processed/Lebreton2017_plastic_emissions.csv')
```
