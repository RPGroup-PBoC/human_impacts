```python
import pandas as pd
import numpy as np
```

```python
old_method = pd.read_csv('source/FAOSTAT_calorie_supply.csv')
old_method.head()
```

```python
new_method = pd.read_csv('source/FAOSTAT_calorie_supply_new.csv')
new_method.head()
```

```python
tidy_dict = {
    'type' : np.append(old_method['Item'].values, new_method['Item'].values),
    'year' : np.append(old_method['Year'].values, new_method['Year'].values),
    'kcal_capita_day' : np.append(old_method['Value'].values, new_method['Value'].values)
}
tidy_df = pd.DataFrame(tidy_dict)
tidy_df.head()
```

```python
sorted_df = tidy_df.sort_values(by='year')
sorted_df.to_csv('processed/FAOSTAT_calorie_supply.csv')
```
