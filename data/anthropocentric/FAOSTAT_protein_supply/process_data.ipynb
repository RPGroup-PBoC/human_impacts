```python
import numpy as np
import pandas as pd
```

```python
old_method = pd.read_csv('source/FAOSTAT_protein_supply.csv')
old_method.head()
```

```python
new_method = pd.read_csv('source/FAOSTAT_protein_supply_new.csv')
new_method.head()
```

```python
old_method['Item'].unique()
```

```python
new_method['Item'].unique()
```

```python
tidy_dict = {
    'type' : np.append(old_method['Item'].values, new_method['Item'].values),
    'year' : np.append(old_method['Year'].values, new_method['Year'].values),
    'g_capita_day' : np.append(old_method['Value'].values, new_method['Value'].values)
}
tidy_df = pd.DataFrame(tidy_dict)
tidy_df.head()
```

```python
tidy_sorted = tidy_df.sort_values(by='year')
#tidy_sorted = tidy_sorted.sort_values(by='type')
tidy_sorted
```

```python
tidy_sorted.to_csv('processed/FAOSTAT_protein_supply.csv')
```
