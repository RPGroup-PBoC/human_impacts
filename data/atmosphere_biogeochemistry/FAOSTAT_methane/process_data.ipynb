```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('source/FAOSTAT_data_2-4-2021.csv', 
                   usecols=(5, 7, 9, 10, 11)
                  )
data.head()
```

```python
data['Element'].unique()
```

```python
methane_df = data[data['Element']=='Emissions (CH4) (Enteric)']
methane_df.head()
```

```python
grouped_df = methane_df.groupby(by='Year').sum()
grouped_df
```

```python
grouped_df = grouped_df.reset_index()
grouped_df
```

```python
tidy_dict = {
    'year': grouped_df['Year'].values,
    'emissions_10^9_kg' : grouped_df['Value'].values/1000
}
tidy_df = pd.DataFrame(tidy_dict)
```

```python
tidy_df.to_csv('processed/FAOSTAT_methane_enteric_fermentation.csv')
```

```python
data = pd.read_csv('source/FAOSTAT_data_rice_paddies.csv', usecols=(5, 9, 10, 11))
data.head()
```

```python
methane_df = data[data['Element']=='Emissions (CH4) (Rice cultivation)']

grouped_df = methane_df.groupby(by='Year').sum()
grouped_df = grouped_df.reset_index()

grouped_df
```

```python
grouped_df = grouped_df.rename(columns={'Value': 'CH4_Gg'})
```

```python
area_df = data[data['Element']=='Area harvested']

area_grouped = area_df.groupby(by='Year').sum()
area_grouped = area_grouped.reset_index()

area_grouped.head()
```

```python
area_grouped = area_grouped.rename(columns={'Value': 'area_ha'})
```

```python
grouped_df.to_csv('processed/FAOSTAT_methane_rice_emissions.csv')
area_grouped.to_csv('processed/FAOSTAT_rice_area.csv')
```
